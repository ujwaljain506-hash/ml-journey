```python
import pandas as pd
```

```python
df = pd.read_csv("/content/matches.csv")
print("File Loaded")
```

```python
total_matches = pd.concat([df['team1'],df['team2']]).value_counts()
```

```python
total_matches
```

```python
win_percent = (df['winner'].value_counts() / total_matches * 100)
```
