```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/matches.csv")
print("File Loaded")
```

```python
df['winner'].value_counts().plot(kind='bar')
plt.title('Number of matches per team won')
plt.xlabel('Team')
plt.ylabel('Number of matches')
plt.show()
```

```python
top5 = df['winner'].value_counts().head(5)
top5.plot(kind='pie')
plt.title('Win Share - Top 5 Teams')
plt.show()
```

```python
top5=df['winner'].value_counts().head(5)
top5.plot(kind='pie')
plt.ylabel('')
plt.title('Top 5 teams won')
plt.show()
```
